```python
from typing import List

def construct2DArray(original: List[int], m: int, n: int) -> List[List[int]]:
    if len(original) != m * n:
        return []

    ans = [[0] * n for _ in range(m)]

    for i, num in enumerate(original):
        ans[i // n][i % n] = num

    return ans
original = [1, 2, 3, 4]
m = 2
n = 2
result = construct2DArray(original, m, n)
print(result)
```

```python
original = [3,6,7,2]
m = 2
n = 2
result = construct2DArray(original, m, n)
print(result)
```

```python
def arrangeCoins(n):
    row = 0
    sum = 0
    
    for i in range(1, n+1):
        sum += i
        if sum <= n:
            row += 1
        else:
            break
    
    return row
n = 5
print(arrangeCoins(n))
```

```python
n = 12
print(arrangeCoins(n))
```

```python
def sortedSquares(nums):
    n = len(nums)
    result = [0] * n
    left, right = 0, n - 1

    for i in range(n - 1, -1, -1):
        if abs(nums[left]) > abs(nums[right]):
            result[i] = nums[left] ** 2
            left += 1
        else:
            result[i] = nums[right] ** 2
            right -= 1

    return result
nums = [-4, -1, 0, 3, 10]
print(sortedSquares(nums))

```

```python
nums = [-7, -1, 0, 4, 10]
print(sortedSquares(nums))

```

```python
def findDisappearedNumbers(nums1, nums2):
    set1 = set(nums1)
    set2 = set(nums2)
    
    result1 = list(set1 - set2)
    result2 = list(set2 - set1)
    
    return [result1, result2]
nums1 = [1, 2, 3]
nums2 = [2, 4, 6]
print(findDisappearedNumbers(nums1, nums2))
```

```python
def findDuplicates(nums):
    result = []

    for num in nums:
        idx = abs(num) - 1

        if nums[idx] > 0:
            nums[idx] = -nums[idx]
        else:
            result.append(abs(num))

    return result

nums = [4, 3, 2, 7, 8, 2, 3, 1]
print(findDuplicates(nums))
```

```python
nums = [1, 2, 3, 4, 5, 6, 6, 7, 8]
print(findDuplicates(nums))
```

```python
def findMin(nums):
    left = 0
    right = len(nums) - 1

    while left < right:
        mid = (left + right) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return nums[left]

nums = [3, 4, 5, 1, 2]
print(findMin(nums))
```

```python
nums = [4, 5, 6, 7, 0, 1, 2]
print(findMin(nums))
```

```python
from collections import defaultdict

def findOriginalArray(changed):
    if len(changed) % 2 != 0:
        return []

    freq = defaultdict(int)
    original = []

    for num in changed:
        freq[num] += 1

    for num in sorted(changed):
        if freq[num] > 0:
            if freq[num * 2] > 0:
                freq[num * 2] -= 1
                original.append(num)
            else:
                return []

    return original
changed = [1, 3, 4, 2, 6, 8]
print(findOriginalArray(changed))
```
