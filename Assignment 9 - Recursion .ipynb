```python
def is_power_of_two(n):
    if n <= 0:
        return False
    return (n & (n - 1)) == 0
print(is_power_of_two(1)) 
print(is_power_of_two(16))  
print(is_power_of_two(3))  
```

```python
print(is_power_of_two(64))  
```

```python
def sum_of_natural_numbers(n):
    sum = 0
    for i in range(1, n+1):
        sum += i
    return sum
print(sum_of_natural_numbers(3))  
print(sum_of_natural_numbers(5))  
```

```python
print(sum_of_natural_numbers(10))  
```

```python
def factorial(N):
    result = 1
    for i in range(1, N+1):
        result *= i
    return result
print(factorial(5))  
print(factorial(4))  
```

```python
print(factorial(10))
```

```python
def calculate_exponent(N, P):
    result = N ** P
    return result
print(calculate_exponent(5, 2))  
print(calculate_exponent(2, 5))  
```

```python
print(calculate_exponent(3, 4))  
```

```python
def find_max(arr):
    # Base case: If the array has only one element, return it
    if len(arr) == 1:
        return arr[0]
    
    # Recursive case: Divide the array into two halves
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    
    # Recursive calls to find the maximum of the left and right subarrays
    max_left = find_max(left)
    max_right = find_max(right)
    
    # Return the larger value between the maximum of the left and right subarrays
    return max(max_left, max_right)
arr1 = [1, 4, 3, -5, -4, 8, 6]
print(find_max(arr1)) 

arr2 = [1, 4, 45, 6, 10, -8]
print(find_max(arr2))  
```

```python
arr3 = [10, 20, 30, 40, 50]
print(find_max(arr3))  
```

```python
def find_nth_term(a, d, N):
    return a + (N - 1) * d
print(find_nth_term(2, 1, 5))  
print(find_nth_term(5, 2, 10))  
```

```python
print(find_nth_term(3, -2, 7))  
```

```python
def permute_string(string, left, right, result):
    if left == right:
        result.append("".join(string))
    else:
        for i in range(left, right + 1):
            string[left], string[i] = string[i], string[left]  # Swap characters
            permute_string(string, left + 1, right, result)  # Recursive call
            string[left], string[i] = string[i], string[left]  # Revert the swap

def print_permutations(S):
    result = []
    string = list(S)
    n = len(string)
    permute_string(string, 0, n - 1, result)
    return result
S1 = "ABC"
print(print_permutations(S1))  

S2 = "XY"
print(print_permutations(S2))  
S1 = "ABC"
print(print_permutations(S1))  
S2 = "XY"
print(print_permutations(S2))  
```

```python
S3 = "ABCD"
print(print_permutations(S3))
```

```python
def product_of_array_elements(arr1):
    product = 1
    for num in arr1:
        product *= num
    return product
arr1 = [1, 2, 3, 4, 5]
print(product_of_array_elements(arr1))  

arr2 = [1, 6, 3]
print(product_of_array_elements(arr2))  
```
