```python
def minimumDeleteSum(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Fill in the first row (deleting characters from s2)
    for j in range(1, n + 1):
        dp[0][j] = dp[0][j - 1] + ord(s2[j - 1])

    # Fill in the first column (deleting characters from s1)
    for i in range(1, m + 1):
        dp[i][0] = dp[i - 1][0] + ord(s1[i - 1])

    # Fill in the rest of the dp array
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(dp[i - 1][j] + ord(s1[i - 1]), dp[i][j - 1] + ord(s2[j - 1]))

    return dp[m][n]
s1 = "sea"
s2 = "eat"
print(minimumDeleteSum(s1, s2))  

```

```python
def isValid(s):
    stack = []
    for char in s:
        if char == '(' or char == '*':
            stack.append(char)
        elif char == ')':
            if stack and stack[-1] == '(':
                stack.pop()
            elif stack and stack[-1] == '*':
                stack.pop()
                if stack and stack[-1] == '(':
                    stack.pop()
            else:
                return False
    return len(stack) == 0
s = "()"
print(isValid(s)) 
```

```python
def minSteps(word1, word2):
    m, n = len(word1), len(word2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        dp[i][0] = i

    for j in range(1, n + 1):
        dp[0][j] = j

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(dp[i - 1][j], dp[i][j - 1]) + 1

    return dp[m][n]
word1 = "sea"
word2 = "eat"
print(minSteps(word1, word2))  
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def constructBinaryTree(s):
    if not s:
        return None

    # Find the root value
    i = 0
    while i < len(s) and (s[i] != '(' and s[i] != ')'):
        i += 1

    root_val = int(s[:i])
    root = TreeNode(root_val)

    # Find the left child subtree
    left_start = i
    count = 0
    while i < len(s):
        if s[i] == '(':
            count += 1
        elif s[i] == ')':
            count -= 1

        if count == 0:
            break

        i += 1

    left_end = i
    root.left = constructBinaryTree(s[left_start + 1: left_end])

    # Find the right child subtree
    if left_end < len(s) - 1:
        root.right = constructBinaryTree(s[left_end + 2: -1])

    return root
s = "4(2(3)(1))(6(5))"
tree = constructBinaryTree(s)

# The tree in pre-order traversal
def preOrderTraversal(node):
    if node is None:
        return

    print(node.val, end=" ")
    preOrderTraversal(node.left)
    preOrderTraversal(node.right)

preOrderTraversal(tree)
```

```python
def compress(chars):
    write = 0
    read = 0
    n = len(chars)

    while read < n:
        curr_char = chars[read]
        count = 1

        while read + 1 < n and chars[read + 1] == curr_char:
            read += 1
            count += 1

        chars[write] = curr_char
        write += 1

        if count > 1:
            count_str = str(count)
            for digit in count_str:
                chars[write] = digit
                write += 1

        read += 1

    return write
chars = ["a","a","b","b","c","c","c"]
new_length = compress(chars)
compressed_chars = chars[:new_length]

print(new_length)         
print(compressed_chars)    
```

```python
from collections import Counter

def findAnagrams(s, p):
    result = []
    p_freq = Counter(p)
    window_freq = Counter(s[:len(p)])

    if window_freq == p_freq:
        result.append(0)

    for i in range(len(p), len(s)):
        outgoing_char = s[i - len(p)]
        incoming_char = s[i]

        window_freq[outgoing_char] -= 1
        if window_freq[outgoing_char] == 0:
            del window_freq[outgoing_char]

        window_freq[incoming_char] += 1

        if window_freq == p_freq:
            result.append(i - len(p) + 1)

    return result
s = "cbaebabacd"
p = "abc"
print(findAnagrams(s, p))  
```

```python
def decodeString(s):
    stack = []
    current_str = ""
    k = 0

    for c in s:
        if c.isdigit():
            k = k * 10 + int(c)
        elif c == "[":
            stack.append((k, current_str))
            k = 0
            current_str = ""
        elif c == "]":
            count, prev_str = stack.pop()
            current_str = prev_str + count * current_str
        else:
            current_str += c

    return current_str
s = "3[a]2[bc]"
print(decodeString(s))  
```

```python
s = "2[abc]3[cd]ef"
print(decodeString(s))
```

```python
def buddyStrings(s, goal):
    if len(s) != len(goal):
        return False

    differing_positions = []
    for i in range(len(s)):
        if s[i] != goal[i]:
            differing_positions.append(i)

    if len(differing_positions) != 2:
        return False

    i, j = differing_positions
    return s[i] == goal[j] and s[j] == goal[i]
s = "ab"
goal = "ba"
print(buddyStrings(s, goal))  
```
