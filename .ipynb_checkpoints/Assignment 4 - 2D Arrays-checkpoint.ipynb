### Assignment 4 - 2D Arrays

```python
def findCommonElements(arr1, arr2, arr3):
    ptr1 = ptr2 = ptr3 = 0
    result = []

    while ptr1 < len(arr1) and ptr2 < len(arr2) and ptr3 < len(arr3):
        if arr1[ptr1] == arr2[ptr2] == arr3[ptr3]:
            result.append(arr1[ptr1])
            ptr1 += 1
            ptr2 += 1
            ptr3 += 1
        elif arr1[ptr1] <= arr2[ptr2] and arr1[ptr1] <= arr3[ptr3]:
            ptr1 += 1
        elif arr2[ptr2] <= arr1[ptr1] and arr2[ptr2] <= arr3[ptr3]:
            ptr2 += 1
        elif arr3[ptr3] <= arr1[ptr1] and arr3[ptr3] <= arr2[ptr2]:
            ptr3 += 1

    return result
arr1 = [1, 2, 3, 4, 5]
arr2 = [1, 2, 5, 7, 9]
arr3 = [1, 3, 4, 5, 8]
print(findCommonElements(arr1, arr2, arr3))

```

```python
def findDisappearedNumbers(nums1, nums2):
    set1 = set(nums1)
    set2 = set(nums2)
    
    result1 = list(set1 - set2)
    result2 = list(set2 - set1)
    
    return [result1, result2]
nums1 = [1, 2, 3]
nums2 = [2, 4, 6]
print(findDisappearedNumbers(nums1, nums2))
```

```python
nums1 = [2, 4, 6, 8, 10]
nums2 = [1, 3, 5, 7, 9]
print(findDisappearedNumbers(nums1, nums2))
```

```python
def transpose(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    
    result = [[0] * rows for _ in range(cols)]
    
    for i in range(rows):
        for j in range(cols):
            result[j][i] = matrix[i][j]
    
    return result
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(transpose(matrix))
```

```python
matrix = [[1, 2, 3, 4], [5, 6, 7, 8]]
print(transpose(matrix))
```

```python
def arrayPairSum(nums):
    nums.sort()
    max_sum = 0
    for i in range(0, len(nums), 2):
        max_sum += nums[i]
    return max_sum
nums = [1, 4, 3, 2]
print(arrayPairSum(nums))
```

```python
nums = [7, 3, 9, 1, 6, 5]
print(arrayPairSum(nums))
```

```python
def arrangeCoins(n):
    row = 0
    sum = 0
    
    for i in range(1, n+1):
        sum += i
        if sum <= n:
            row += 1
        else:
            break
    
    return row
n = 5
print(arrangeCoins(n))
```

```python
n = 11
print(arrangeCoins(n))
```

```python
def sortedSquares(nums):
    n = len(nums)
    result = [0] * n
    left, right = 0, n - 1

    for i in range(n - 1, -1, -1):
        if abs(nums[left]) > abs(nums[right]):
            result[i] = nums[left] ** 2
            left += 1
        else:
            result[i] = nums[right] ** 2
            right -= 1

    return result
nums = [-4, -1, 0, 3, 10]
print(sortedSquares(nums))
```

```python
nums = [-7, -3, -1, 4, 8]
print(sortedSquares(nums))
```

```python
def maxCount(m, n, ops):
    min_row = m
    min_col = n

    for op in ops:
        min_row = min(min_row, op[0])
        min_col = min(min_col, op[1])

    return min_row * min_col
m = 3
n = 3
ops = [[2, 2], [3, 3]]
print(maxCount(m, n, ops))
```

```python
m = 4
n = 5
ops = [[3, 2], [2, 3], [4, 1]]
print(maxCount(m, n, ops))
```

```python
def rearrange(nums, n):
    x = nums[:n]
    y = nums[n:]
    result = []

    for i in range(n):
        result.append(x[i])
        result.append(y[i])

    return result
nums = [2, 5, 1, 3, 4, 7]
n = 3
print(rearrange(nums, n))
```

```python
nums = [1,2,3,4,5,6]
n = 3
print(rearrange(nums, n))
```
