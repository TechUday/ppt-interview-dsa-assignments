### Assignment 11 - Binary Search

```python
def mySqrt(x):
    if x == 0:
        return 0

    left = 1
    right = x

    while left <= right:
        mid = left + (right - left) // 2
        if mid * mid > x:
            right = mid - 1
        else:
            left = mid + 1

    return right
print(mySqrt(16))  
```

```python
print(mySqrt(25))  
```

```python
def findPeakElement(nums):
    left = 0
    right = len(nums) - 1

    while left < right:
        mid = left + (right - left) // 2

        if nums[mid] < nums[mid + 1]:
            left = mid + 1
        else:
            right = mid

    return left
print(findPeakElement([1, 2, 3, 1]))
```

```python
print(findPeakElement([1, 2, 1, 3, 5, 6, 4]))  
```

```python
def missingNumber(nums):
    missing = len(nums)
    for i in range(len(nums)):
        missing ^= i ^ nums[i]
    return missing
print(missingNumber([3, 0, 1]))  
```

```python
print(missingNumber([0,1]))
```

```python
def findDuplicate(nums):
    slow = nums[0]
    fast = nums[0]

    while True:
        slow = nums[slow]
        fast = nums[nums[fast]]
        if slow == fast:
            break

    slow = nums[0]
    while slow != fast:
        slow = nums[slow]
        fast = nums[fast]

    return slow
print(findDuplicate([1, 3, 4, 2, 2]))  
```

```python
print(findDuplicate([3,1,3,4,2]))  
```

```python
def intersection(nums1, nums2):
    set1 = set(nums1)
    set2 = set(nums2)
    return list(set1.intersection(set2))
print(intersection([1, 2, 2, 1], [2, 2]))  
```

```python
print(intersection([4,9,5], [9,4,9,8,4]))  
```

```python
def findMin(nums):
    left = 0
    right = len(nums) - 1

    while left < right:
        mid = left + (right - left) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return nums[left]
print(findMin([4, 5, 6, 7, 0, 1, 2]))  
```

```python
print(findMin([3,4,5,1,2]))  
```

```python
def searchRange(nums, target):
    left = findPosition(nums, target, True)
    right = findPosition(nums, target, False)

    return [left, right]

def findPosition(nums, target, isLeft):
    position = -1
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] == target:
            position = mid
            if isLeft:
                right = mid - 1
            else:
                left = mid + 1
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return position
print(searchRange([5, 7, 7, 8, 8, 10], 8))  
```

```python
print(searchRange([5, 7, 7, 8, 8, 10], 6))  
```

```python
from collections import Counter

def intersect(nums1, nums2):
    counter1 = Counter(nums1)
    result = []

    for num in nums2:
        if num in counter1 and counter1[num] > 0:
            result.append(num)
            counter1[num] -= 1

    return result
print(intersect([1, 2, 2, 1], [2, 2]))  
```

```python
print(intersect([4,9,5], [9,4,9,8,4]))  
```
