```python
def arrayPairSum(nums):
    nums.sort()  # Sort the array in ascending order
    total = 0
    for i in range(0, len(nums), 2):
        total += nums[i]  # Add the smaller element of each pair
    return total


nums = [1, 4, 3, 2]
result = arrayPairSum(nums)
print(result) 
```

```python

nums = [5, 10, 2, 7, 15, 3]
result = arrayPairSum(nums)
print(result) 
```

```python
def maxCandies(candyType):
    unique_types = len(set(candyType))  # Count the number of unique candy types
    max_candies = min(unique_types, len(candyType) // 2)  # Consider the minimum between unique types and n / 2
    return max_candies


candyType = [1, 1, 2, 2, 3, 3]
result = maxCandies(candyType)
print(result)  
```

```python

candyType = [1, 1, 2, 3, 3, 4, 4, 5]
result = maxCandies(candyType)
print(result)  
```

```python
from collections import Counter

def findLHS(nums):
    num_counts = Counter(nums)  # Count occurrences of each number
    max_length = 0

    for num in num_counts:
        if num + 1 in num_counts:
            length = num_counts[num] + num_counts[num + 1]
            max_length = max(max_length, length)

    return max_length


nums = [1, 3, 2, 2, 5, 2, 3, 7]
result = findLHS(nums)
print(result) 
```

```python

nums = [1, 2, 3, 4, 5, 6, 7, 8]
result = findLHS(nums)
print(result)  
```

```python
def canPlaceFlowers(flowerbed, n):
    length = len(flowerbed)
    count = 0
    i = 0

    while i < length:
        if flowerbed[i] == 0 and (i == 0 or flowerbed[i - 1] == 0) and (i == length - 1 or flowerbed[i + 1] == 0):
            flowerbed[i] = 1  # Plant a flower
            count += 1
        i += 1

    return count >= n


flowerbed = [1, 0, 0, 0, 1]
n = 1
result = canPlaceFlowers(flowerbed, n)
print(result)  
```

```python

flowerbed = [0, 0, 0, 0, 0]
n = 3
result = canPlaceFlowers(flowerbed, n)
print(result)  
```

```python
def maximumProduct(nums):
    nums.sort()  # Sort the array in ascending order
    n = len(nums)
    return max(nums[n - 1] * nums[n - 2] * nums[n - 3], nums[0] * nums[1] * nums[n - 1])


nums = [1, 2, 3]
result = maximumProduct(nums)
print(result)  
```

```python

nums = [-1, -2, -3, -4, -5]
result = maximumProduct(nums)
print(result)  
```

```python
def search(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1


nums = [-1, 0, 3, 5, 9, 12]
target = 9
result = search(nums, target)
print(result)
```

```python

nums = [-1, 0, 3, 5, 9, 12]
target = 2
result = search(nums, target)
print(result)  
```

```python
def isMonotonic(nums):
    is_increasing = True
    is_decreasing = True

    for i in range(1, len(nums)):
        if nums[i] < nums[i - 1]:
            is_increasing = False
        if nums[i] > nums[i - 1]:
            is_decreasing = False

    return is_increasing or is_decreasing


nums = [6, 5, 4, 4]
result = isMonotonic(nums)
print(result) 
```

```python

nums = [1, 2, 4, 4, 5, 6, 7]
result = isMonotonic(nums)
print(result) 
```

```python
def minimumScore(nums, k):
    min_num = min(nums)
    max_num = max(nums)
    
    if max_num - min_num <= 2 * k:
        return 0

    # Scenario 1: Decrease the maximum element by k
    potential_score_1 = max_num - k

    # Scenario 2: Increase the minimum element by k
    potential_score_2 = min_num + k

    return min(potential_score_1, potential_score_2)


nums = [1]
k = 0
result = minimumScore(nums, k)
print(result)  
```

```python

nums = [1, 2, 3, 4, 5]
k = 2
result = minimumScore(nums, k)
print(result) 
```
