```python
def is_power_of_three(n):
    if n <= 0:
        return False
    while n % 3 == 0:
        n /= 3
    return n == 1
print(is_power_of_three(27))  
```

```python
print(is_power_of_three(45))  
```

```python
def last_remaining_number(n):
    if n == 1:
        return 1
    
    if n % 2 == 1:
        return 2 * last_remaining_number(n // 2)
    else:
        return 2 * last_remaining_number(n // 2) - 1
print(last_remaining_number(9))  
```

```python
print(last_remaining_number(15))
```

```python
def print_subsets(s, subset="", index=0):
    if index == len(s):
        print(subset)
        return
    
    print_subsets(s, subset, index + 1)  # Exclude current character
    print_subsets(s, subset + s[index], index + 1)  # Include current character

# Driver code
s = "abc"
print_subsets(s)
set1 = "abc"
print_subsets(set1)

set2 = "abcd"
print_subsets(set2)
```

```python
set3 = "xy"
print_subsets(set3)
```

```python
def string_length(s):
    if s == "":
        return 0
    return 1 + string_length(s[1:])


print(string_length("abc")) 
print(string_length("Hello, World!")) 
```

```python
def count_contiguous_substrings(S):
    count = 0
    n = len(S)
    for i in range(n):
        for j in range(i, n):
            if S[i] == S[j]:
                count += 1
    return count


print(count_contiguous_substrings("abc"))  
print(count_contiguous_substrings("aba"))  
```

```python
print(count_contiguous_substrings("aabbaa"))  
```

```python
def tower_of_hanoi(n, source, destination, auxiliary):
    if n == 1:
        print(f"Move disk 1 from {source} to {destination}")
        return 1
    else:
        moves = tower_of_hanoi(n - 1, source, auxiliary, destination)
        print(f"Move disk {n} from {source} to {destination}")
        moves += 1
        moves += tower_of_hanoi(n - 1, auxiliary, destination, source)
        return moves


n = 3  # Number of discs
source = "A"  # Source rod
destination = "C"  # Destination rod
auxiliary = "B"  # Auxiliary rod

total_moves = tower_of_hanoi(n, source, destination, auxiliary)
print("Total moves:", total_moves)
n = 2
source = "A"
destination = "C"
auxiliary = "B"

total_moves = tower_of_hanoi(n, source, destination, auxiliary)
print("Total moves:", total_moves)
```

```python
from itertools import permutations

def print_permutations(string):
    perms = permutations(string)
    for perm in perms:
        print(''.join(perm))

# Example usage
string1 = "cd"
print("Permutations of", string1 + ":")
print_permutations(string1)

string2 = "abb"
print("Permutations of", string2 + ":")
print_permutations(string2)
```

```python
str3 = "abc"
print("Permutations of", str3 + ":")
print_permutations(str3)

```

```python
def count_consonants(string):
    vowels = "aeiou"
    count = 0
    for ch in string:
        if ch.isalpha() and ch.lower() not in vowels:
            count += 1
    return count


string1 = "Hello World"
consonant_count1 = count_consonants(string1)
print("Total consonants in the string:", consonant_count1)

string2 = "Python Programming"
consonant_count2 = count_consonants(string2)
print("Total consonants in the string:", consonant_count2)
```
